# ORB Strategy — Quick Data Viewer

Lightweight notebook to browse ObjectStore files and download CSVs.

```python
from AlgorithmImports import *
import pandas as pd
import io
from IPython.display import HTML
import base64

qb = QuantBook()

# List all ObjectStore keys
try:
    all_keys = sorted([str(k) for k in qb.object_store.get_file_paths()])
except:
    all_keys = sorted(list(qb.object_store.keys))

print(f'ObjectStore has {len(all_keys)} files:\n')
for k in all_keys:
    print(f'  {k}')
```

```python
# Load and display any CSV from ObjectStore
# Change KEY to view different files
KEY = 'trade_log.csv'

if qb.object_store.contains_key(KEY):
    data = qb.object_store.read(KEY)
    df = pd.read_csv(io.StringIO(data))
    print(f'Loaded {len(df)} rows, {len(df.columns)} columns from: {KEY}\n')
    display(df)

    # Download link
    csv_str = df.to_csv(index=False)
    b64 = base64.b64encode(csv_str.encode()).decode()
    display(HTML(f'<a download="{KEY}" href="data:text/csv;base64,{b64}">Download {KEY}</a>'))
else:
    print(f'Key not found: {KEY}')
```
